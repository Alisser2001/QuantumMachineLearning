```python
# ============================================================
# CQ-Optuna-NN-Kernel v1: NN con "capa kernel cuántico" (landmarks)
# Por cada (modo × n_qubits): Optuna con 3 trials
# ============================================================
import os, time, requests, numpy as np, pandas as pd, matplotlib.pyplot as plt, optuna, torch
from torch import nn
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import LocalOutlierFactor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from qiskit import QuantumCircuit
from qiskit.circuit.library import ZZFeatureMap, PauliFeatureMap
from qiskit.quantum_info import Statevector, SparsePauliOp

# ---------------- CONFIG ----------------
SEED = 42
np.random.seed(SEED)
torch.manual_seed(SEED)

RESULTS_DIR = "cq_v10_nn_kernel_outputs"
os.makedirs(f"{RESULTS_DIR}/figs", exist_ok=True)
os.makedirs(f"{RESULTS_DIR}/circuits", exist_ok=True)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
QUBITS_LIST = [4, 6, 8]
MODES = ["Z", "Z+ZZ", "Pauli", "Angle", "Fourier", "RandomizedPauli"]
N_TRIALS = 3
BATCH_SIZE = 32
EPOCHS = 50  # entreno breve

# ---------------- DATASET ----------------
url = "https://raw.githubusercontent.com/Alisser2001/QuantumMachineLearning/emmadev/concrete.dat"
resp = requests.get(url)
lines = [line.strip() for line in resp.text.splitlines()]
attributes = [line.split()[1] for line in lines if line.startswith("@attribute")]
data_start = lines.index("@data") + 1
data = [line.split(",") for line in lines[data_start:] if line]
df = pd.DataFrame(data, columns=attributes).astype(float)

X_full = df.drop(columns=["ConcreteCompressiveStrength"]).values
y_full = df["ConcreteCompressiveStrength"].values
mask = LocalOutlierFactor(n_neighbors=20, contamination=0.05).fit_predict(
    StandardScaler().fit_transform(X_full)
) == 1
X, y = X_full[mask], y_full[mask]
print(f"[LOF] Muestras: {len(df)} → {len(X)} (removidos {len(df)-len(X)})")

X_tr, X_te, y_tr, y_te = train_test_split(X, np.log1p(y), test_size=0.2, random_state=SEED)

# ---------------- QUANTUM PART: feature map y estado ----------------
def make_circuit(x, n, mode):
    if mode == "Z":
        qc = QuantumCircuit(n)
        for i, val in enumerate(x[:n]): qc.rz(float(val), i)
        return qc
    if mode == "Z+ZZ":
        fm = ZZFeatureMap(feature_dimension=n, reps=2, entanglement="full")
        return fm.assign_parameters(dict(zip(fm.parameters, x[:n])))
    if mode == "Pauli":
        fm = PauliFeatureMap(feature_dimension=n, reps=2, paulis=["X","Y","Z"])
        return fm.assign_parameters(dict(zip(fm.parameters, x[:n])))
    if mode == "Angle":
        qc = QuantumCircuit(n)
        for i, val in enumerate(x[:n]): qc.ry(float(val), i)
        for i in range(n-1): qc.cz(i, i+1)
        return qc
    if mode == "Fourier":
        qc = QuantumCircuit(n)
        for i, val in enumerate(x[:n]):
            qc.h(i); qc.rz(2*np.pi*float(val), i)
        for i in range(n-1): qc.cz(i, i+1)
        return qc
    if mode == "RandomizedPauli":
        qc = QuantumCircuit(n); rng = np.random.default_rng(SEED)
        for i, val in enumerate(x[:n]): qc.ry(float(val), i)
        for _ in range(2):
            for i in range(n): getattr(qc, rng.choice(["x", "y", "z"]))(i)
            for i in range(n-1): qc.cz(i, i+1)
        return qc
    raise ValueError(mode)

def states_matrix(X_pca, n_qubits, mode):
    """Devuelve una matriz compleja M de tamaño (N, 2^n) con los vectores de estado.
       También devuelve qc_sample para métricas de circuito."""
    N = len(X_pca); dim = 2**n_qubits
    M = np.empty((N, dim), dtype=np.complex128)
    qc_sample = None
    for i, x in enumerate(X_pca):
        qc = make_circuit(x, n_qubits, mode)
        if qc_sample is None: qc_sample = qc
        sv = Statevector.from_instruction(qc).data  # (2^n,)
        M[i, :] = sv
    return M, qc_sample

def circuit_metrics(qc):
    return {
        "depth": qc.depth(),
        "size": qc.size(),
        "multi_qubit_gates": sum(1 for g in qc.data if len(g.qubits) > 1),
        "num_qubits": qc.num_qubits,
        "ops": qc.count_ops()
    }

# ---------------- NN ----------------
class SimpleRegressor(nn.Module):
    def __init__(self, in_dim, h1, h2, dropout):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, h1), nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(h1, h2), nn.ReLU(),
            nn.Linear(h2, 1)
        )
    def forward(self, x): return self.net(x)

def train_nn(model, train_loader, X_val, y_val, epochs, lr):
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = nn.MSELoss()
    model.train()
    for _ in range(epochs):
        for xb, yb in train_loader:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            pred = model(xb).squeeze()
            loss = loss_fn(pred, yb)
            opt.zero_grad(); loss.backward(); opt.step()
    model.eval()
    with torch.no_grad():
        y_pred = model(torch.tensor(X_val, dtype=torch.float32).to(DEVICE)).cpu().numpy().squeeze()
    return y_pred

# ---------------- LOOP PRINCIPAL: por (modo × n_qubits) ----------------
records = []

for n_qubits in QUBITS_LIST:
    # Preprocesamiento clásico fijo por n_qubits
    scaler = StandardScaler().fit(X_tr)
    Xtr_s, Xte_s = scaler.transform(X_tr), scaler.transform(X_te)
    pca = PCA(n_components=n_qubits, random_state=SEED).fit(Xtr_s)
    Xtr_p, Xte_p = pca.transform(Xtr_s), pca.transform(Xte_s)

    # Precomputo matrices de estado para reusar dentro de cada modo
    # (para ahorrar tiempo al variar landmarks en Optuna)
    # Las matrices dependen del modo, así que las calculo dentro del ciclo de modo.

    for mode in MODES:
        print(f"\n=== Optimizando NN+QKernel: {mode} con {n_qubits} qubits ===")

        # Construyo matrices de estado COMPLETAS (train/test) para este modo
        M_tr, qc_sample = states_matrix(Xtr_p, n_qubits, mode)  # (Ntr, 2^n)
        M_te, _        = states_matrix(Xte_p, n_qubits, mode)  # (Nte, 2^n)
        circ_info = circuit_metrics(qc_sample)

        # Índices barajados del train para landmarks deterministas
        rng = np.random.default_rng(SEED)
        perm_tr = rng.permutation(M_tr.shape[0])

        # Optuna (3 trials) por esta configuración
        def objective(trial):
            # Nº de landmarks (tamaño del diccionario del kernel)
            m = trial.suggest_categorical("landmarks", [16, 32, 64])

            # Hiperparámetros de la MLP
            h1 = trial.suggest_int("hid1", 16, 128, log=True)
            h2 = trial.suggest_int("hid2", 8, 64, log=True)
            dropout = trial.suggest_float("dropout", 0.0, 0.4)
            lr = trial.suggest_float("lr", 1e-4, 1e-2, log=True)

            # Selección de landmarks (primeros m de la permutación fija)
            L_idx = perm_tr[:m]
            L = M_tr[L_idx, :]  # (m, d)

            # Construyo feature kernel: K(x) = |⟨ψ(x), ψ(L_j)⟩|^2  para j=1..m
            # Vectorizado: K = | M @ L^H |^2
            K_tr = M_tr @ L.conj().T
            K_te = M_te @ L.conj().T
            K_tr = np.abs(K_tr) ** 2
            K_te = np.abs(K_te) ** 2

            # Escalo las features del kernel (estabiliza el entrenamiento)
            k_scaler = StandardScaler().fit(K_tr)
            Ktr_s = k_scaler.transform(K_tr)
            Kte_s = k_scaler.transform(K_te)

            # Pytorch dataloaders
            ds = TensorDataset(
                torch.tensor(Ktr_s, dtype=torch.float32),
                torch.tensor(y_tr,  dtype=torch.float32)
            )
            dl = DataLoader(ds, batch_size=min(BATCH_SIZE, len(ds)), shuffle=True)

            model = SimpleRegressor(in_dim=m, h1=h1, h2=h2, dropout=dropout).to(DEVICE)
            y_pred = train_nn(model, dl, Kte_s, y_te, EPOCHS, lr)

            mse = mean_squared_error(y_te, y_pred)
            return mse

        study = optuna.create_study(direction="minimize")
        study.optimize(objective, n_trials=N_TRIALS, show_progress_bar=False)

        best = study.best_trial
        rec = {
            "mode": mode,
            "qubits": n_qubits,
            "best_MSE": best.value,
            "best_landmarks": best.params["landmarks"],
            "best_hid1": best.params["hid1"],
            "best_hid2": best.params["hid2"],
            "best_dropout": best.params["dropout"],
            "best_lr": best.params["lr"],
            "depth": circ_info["depth"],
            "multi_qubit_gates": circ_info["multi_qubit_gates"]
        }
        records.append(rec)
        print(f"→ Mejor trial: MSE={rec['best_MSE']:.6f}, "
              f"m={rec['best_landmarks']}, h1={rec['best_hid1']}, "
              f"h2={rec['best_hid2']}, lr={rec['best_lr']:.2e}")

# ---------------- RESULTADOS ----------------
df = pd.DataFrame(records)
csv_path = f"{RESULTS_DIR}/cq_v10_nn_kernel_best.csv"
df.to_csv(csv_path, index=False)
print("\n[OK] Resultados guardados:", csv_path)
print(df[["mode","qubits","best_MSE","best_landmarks","best_hid1","best_hid2","depth","multi_qubit_gates"]])

# ---------------- GRÁFICA ----------------
plt.figure(figsize=(10,6))
for m in MODES:
    sub = df[df["mode"]==m].sort_values("qubits")
    if not sub.empty:
        plt.plot(sub["qubits"], sub["best_MSE"], marker='o', label=m)
plt.title("Mejor MSE por modo y nº de qubits (NN con kernel cuántico)")
plt.xlabel("Qubits"); plt.ylabel("MSE"); plt.grid(True); plt.legend(); plt.tight_layout()
plt.savefig(f"{RESULTS_DIR}/figs/mse_vs_qubits_nn_qkernel.png", dpi=160)
plt.show()
```
