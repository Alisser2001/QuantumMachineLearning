```python
# =============================================
# âš›ï¸ QSVR con IBM Quantum Runtime (versiÃ³n moderna Qiskit 1.2+)
# =============================================

# Si estÃ¡s en Google Colab o entorno limpio, instala primero:
# !pip install qiskit qiskit-ibm-runtime qiskit-machine-learning qiskit_algorithms --quiet

from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit.circuit.library import ZZFeatureMap
from qiskit_machine_learning.algorithms import QSVR
from qiskit_algorithms.utils import algorithm_globals
from qiskit.quantum_info import Statevector

import pandas as pd
import numpy as np
import requests
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import LocalOutlierFactor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import time

# =============================================
# ğŸ”‘ ConexiÃ³n con IBM Quantum (IBM Cloud)
# =============================================

# Elimina credenciales previas (si existen)
try:
    QiskitRuntimeService.delete_account()
except Exception:
    pass

# Guarda tu cuenta (solo necesitas hacerlo una vez)
QiskitRuntimeService.save_account(
    channel="ibm_cloud",
    token="   ",  # tu token IBM Cloud
    instance="crn:v1:bluemix:public:quantum-computing:us-east:a/86afa610efa6479e8fc1d3f61ec0c301:d67e13de-a56b-4169-95cf-5626b3daaf1a::",  # <-- reemplaza con tu CRN real
    overwrite=True
)

# Carga el servicio y selecciona el backend
service = QiskitRuntimeService(channel="ibm_cloud")

backends = service.backends()
print("\nğŸ” Backends disponibles:")
for b in backends:
    print(" -", b.name)

backend = service.least_busy()
print(f"\nâœ… Conectado a IBM Cloud â†’ Backend real: {backend.name}")


# =============================================
# ğŸ“¥ Carga del dataset
# =============================================
url = "https://raw.githubusercontent.com/Alisser2001/QuantumMachineLearning/emmadev/elevators.dat"
resp = requests.get(url)
lines = [l.strip() for l in resp.text.splitlines()]
attributes = [line.split()[1] for line in lines if line.lower().startswith("@attribute")]
data_start = next(i for i, l in enumerate(lines) if l.lower() == "@data") + 1
rows = [ln for ln in lines[data_start:] if ln and not ln.startswith('%')]
data = [r.split(',') for r in rows]
df = pd.DataFrame(data, columns=attributes).apply(pd.to_numeric)

features = ['SaTime1', 'SaTime2', 'SaTime3', 'Sa', 'SaTime4', 'AbsRoll', 'Q', 'ClimbRate']
target = 'Goal'
X, y = df[features], df[target]

# =============================================
# ğŸ§¹ Preprocesamiento
# =============================================
lof = LocalOutlierFactor(n_neighbors=10, contamination=0.05)
mask = lof.fit_predict(X) == 1
X, y = X[mask], y[mask]

for c in ['SaTime1', 'SaTime2', 'SaTime3', 'Sa', 'SaTime4']:
    X[c] = np.log1p(np.clip(X[c].values, a_min=0, a_max=None))

scaler = MinMaxScaler(feature_range=(-np.pi, np.pi))
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
print("âœ… Datos preparados:", X_train.shape)

# =============================================
# âš›ï¸ ConstrucciÃ³n del kernel cuÃ¡ntico manual
# =============================================
algorithm_globals.random_seed = 42
num_qubits = X_train.shape[1]
feature_map = ZZFeatureMap(feature_dimension=num_qubits, reps=2, entanglement="linear")

def compute_kernel_matrix(X1, X2):
    """Calcula la matriz de kernel de fidelidad entre conjuntos de datos."""
    K = np.zeros((len(X1), len(X2)))
    for i, x1 in enumerate(X1):
        psi1 = Statevector(feature_map.assign_parameters(x1))
        for j, x2 in enumerate(X2):
            psi2 = Statevector(feature_map.assign_parameters(x2))
            K[i, j] = np.abs(psi1.inner(psi2)) ** 2
    return K

# Calcula las matrices de kernel
print("\nâš™ï¸ Calculando matrices de kernel...")
K_train = compute_kernel_matrix(X_train, X_train)
K_test = compute_kernel_matrix(X_test, X_train)

# =============================================
# ğŸš€ Entrenamiento del modelo QSVR
# =============================================
print("\nğŸš€ Entrenando modelo QSVR...")
start_time = time.time()

qsvr = QSVR(kernel="precomputed", C=2.0)
qsvr.fit(K_train, y_train)

train_time = time.time() - start_time
print(f"âœ… Entrenamiento completado en {train_time:.2f} s")

# =============================================
# ğŸ“Š EvaluaciÃ³n del modelo
# =============================================
y_pred = qsvr.predict(K_test)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nğŸ“ˆ RESULTADOS (IBM Cloud Runtime):")
print(f"MSE = {mse:.6f}")
print(f"MAE = {mae:.6f}")
print(f"RÂ²  = {r2:.4f}")

# =============================================
# ğŸ“‰ GrÃ¡fica
# =============================================
plt.figure(figsize=(7,5))
plt.scatter(y_test, y_pred, color='darkviolet', alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.title("QSVR - IBM Quantum (Estado vectorial)")
plt.xlabel("Valor real")
plt.ylabel("PredicciÃ³n")
plt.grid(True)
plt.show()
```
